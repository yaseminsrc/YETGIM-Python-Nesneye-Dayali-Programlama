```python
#Modül 1 Örnekler

#Soru: Kullanıcıdan dairenin yarıçapını alın. Çevre ve alanını hesaplayın.


# çevre = 2 * pi * r      alan =  pi * r^2
import math

r = float(input("Yarıçap giriniz : "))
pi = math.pi

cevre = 2 * pi * r
alan = pi * r**2

print(f"Çevre : {cevre}")
print("Alan : {}".format(alan))
```

```python

#Soru: Kullanıcıdan alınan 2 sayıya toplama,çıkarma,çarpma,bölme işlemlerini yapınız.

int_val1 = int(input("Bir sayı giriniz : "))
int_val2 = int(input("Bir sayı giriniz : "))

print(f"Toplam : {int_val1 + int_val2}")
print(f"Çıkarma : {int_val1 - int_val2}")
print(f"Çarpma : {int_val1 * int_val2}")
print(f"Bölme : {int_val1 / int_val2}")
```

```python
#Vücut Kitle İndeksi (BMI)
#Soru: Kullanıcının kilosunu ve boyunu alarak BMI değerini hesaplayın. BMI = Kilo / (Boy * Boy)

kilo = float(input("Kilonuzu girin (kg): "))
boy = float(input("Boyunuzu girin (metre): "))

bmi = kilo / (boy ** 2)

print(f"Vücut Kitle İndeksiniz (BMI): {bmi:.2f}")
```
